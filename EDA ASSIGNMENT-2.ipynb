```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'

# Load the dataset
df = pd.read_csv(url)

# Clean the "Price" column by converting "Everyone" to NaN
df["Price"] = df["Price"].apply(lambda x: None if x == "Everyone" else x)

# Convert the "Price" column to numeric
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")

# Group by Category and calculate mean price
grouped_df = df.groupby("Category")["Price"].mean()

# Create the bar plot
plt.figure(figsize=(12, 6))
plt.bar(grouped_df.index, grouped_df.values)
plt.xlabel("App Category")
plt.ylabel("Average Price (USD)")
plt.title("Average Price of Apps by Category")
plt.xticks(rotation=90)
plt.tight_layout()

# Show the plot
plt.show()
```
