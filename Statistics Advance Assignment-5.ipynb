```python
import scipy.stats as stats
mean = 50
std_dev = 5
confidence_level = 0.95
sample_size = 100  # You need to provide the sample size for accurate confidence interval calculation
standard_error = std_dev / (sample_size ** 0.5)
z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_of_error = z_score * standard_error
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
import scipy.stats as stats
observed_frequencies = [15, 18, 22, 11, 12, 22]  
expected_frequencies = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]  
total_observations = sum(observed_frequencies)
expected_counts = [total_observations * p for p in expected_frequencies]
chi_square_statistic, p_value = stats.chisquare(observed_frequencies, expected_counts)
alpha = 0.05
print(f"Chi-Square Statistic: {chi_square_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The distribution of colors does not match the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The distribution of colors matches the expected distribution.")
```

```python
import math
sample_size = 500
number_of_smokers = 60
confidence_level = 0.95
z_critical = 1.96  # For 95% confidence level
sample_proportion = number_of_smokers / sample_size
standard_error = math.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)
margin_of_error = z_critical * standard_error
lower_bound = sample_proportion - margin_of_error
upper_bound = sample_proportion + margin_of_error
print("95% Confidence Interval:", (lower_bound, upper_bound))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
df = 10
x = np.linspace(0, 30, 500)
chi_squared = stats.chi2.pdf(x, df)
plt.figure(figsize=(8, 6))
plt.plot(x, chi_squared, 'b-', label='Chi-Square Distribution (df=10)')
x_shade = np.linspace(15, 30, 500)
y_shade = stats.chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='gray', alpha=0.3, label='Area for X > 15')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

Null hypothesis: The coin is fair.
Alternative hypothesis: The coin is biased towards tails.
The expected frequencies for a fair coin are 50 heads and 50 tails. The observed frequencies are 45 tails and 55 heads.

We can use the chi-square goodness of fit test to test the hypothesis. The chi-square statistic is calculated as follows:

2 =  (O - E)^2 / E
where O is the observed frequency, E is the expected frequency, and the summation is over all the categories.

In this case, the chi-square statistic is calculated as follows:

2 = (45 - 50)^2 / 50 + (55 - 50)^2 / 50 = 0.0
The degrees of freedom for the chi-square test is 1, since there is only one category (heads or tails).

The critical value for the chi-square test at a significance level of 0.05 is 3.841.

Since the chi-square statistic is equal to 0.0, which is less than the critical value, we cannot reject the null hypothesis.

Alternative hypothesis: Smoking status and lung cancer diagnosis are dependent.
The contingency table shows the frequencies of people with lung cancer (yes or no) and smoking status (smoker or non-smoker).

We can use the chi-square test for independence to test the hypothesis. The chi-square statistic is calculated as follows:

2 =  (O - E)^2 / E
where O is the observed frequency, E is the expected frequency, and the summation is over all the cells in the contingency table.

The expected frequencies can be calculated as follows:

E = (row total * column total) / grand total
In this case, the chi-square statistic is calculated as follows:

2 = (60 - 66.67)^2 / 66.67 + (140 - 127.78)^2 / 127.78 + (30 - 37.5)^2 / 37.5 + (170 - 135)^2 / 135 = 9.23

The degrees of freedom for the chi-square test is (2 - 1)(3 - 1) = 2.

The critical value for the chi-square test at a significance level of 0.05 is 5.991.

Since the chi-square statistic is equal to 9.23, which is greater than the critical value, we can reject the null hypothesis.
